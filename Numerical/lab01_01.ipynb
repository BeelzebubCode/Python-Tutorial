```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
a = 4
b = -2
x = np.arange(1, 1e5)
y = a*x**(b)
```

```python
plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.xscale("log")
plt.yscale("log")
```

```python
"""
np.logspace(start, stop, num)

start=1:  10  10^1
stop=6:  10  10^6
num=6:  10^1 - 10^6
"""
n = np.logspace(1, 6, 100) 


error1 = 1 / n**1
error15 = 1 / n**1.5
error2 = 1 / n**2
```

```python
np.logspace(1, 6, 6)
```

```python
p = plt.plot
p(n, error1, label=r"$n^{-1}$")
p(n, error15, label=r"$n^{-1.5}$")
p(n, error2, label=r"$n^{-2}$")

plt.xlabel("$n$")
plt.ylabel("error")
plt.grid()
plt.legend(frameon=False)
plt.xscale("log")
plt.yscale("log")
```

```python
n = np.logspace(0, 1, 100)

error21 = 2**-n
error23 = 2**-(3*n)
error2e = np.exp(-2*n)
```

```python
p = plt.plot
p(n, error21, label=r"$2^{-n}$")
p(n, error23, label=r"$2^{-3n}$")
p(n, error2e, label=r"$e^{-2n}$")

plt.xlabel("$n$")
plt.ylabel("error")
plt.grid()
plt.legend(frameon=False)
plt.xscale("log")
plt.yscale("log")
```

```python
n = np.logspace(0, 2, 100)

error2e = np.exp(-2*n)
error2 = 1 / n**2

p = plt.plot
p(n, error2e, label=r"$e^{-2n}$")
p(n, error2, label=r"$n^{-2}$")

plt.xlabel("$n$")
plt.ylabel("error")
plt.grid()
plt.legend(frameon=False)
# plt.xscale("log")
plt.yscale("log")
```
