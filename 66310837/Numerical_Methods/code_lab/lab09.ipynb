## Example 1

```python
A = np.array([[0.9,0.5],
              [0.1,0.5]])
labels = ["Sunny","Rainy"]
graph_matrix(A, mat_label=labels, show_weights=True, round_digits=3)
```

The weather today is sunny, What is the probability of sunny day tomorrow?

```python
x0 = np.array([0,1])
x1 = A@x0
x2 = A@x1
x3 = A@x2
x3
```

```python
its = 20
allx = np.zeros((2,its))
```

```python
allx
```

```python
x = np.array([0,1])
allx[:,0] = x
allx
```

```python
for i in range(1,its):
    allx[:,i] = A@allx[:,i-1]
```

```python
allx.T
```

```python
print('Probabilities of initial state:',allx[:,0])
print('Probabilities after 20 days:', allx[:,-1])
plt.plot(allx.T)
plt.xlabel('')
```

```python
a = random.random()
b = 1 - a
allx[:,0] = np.array([a,b])
```

```python
for i in range(1,its):
    allx[:,i] = A@allx[:,i-1]
    
allx.T
```

```python
print('Probabilities of initial state:',allx[:,0])
print('Probabilities after 20 days:', allx[:,-1])
plt.plot(allx.T)
plt.xlabel('')
```

```python
lambdas, U = la.eig(A)
print(lambdas)
print(U[:,0])
```

```python
U[:,0] / la.norm(U[:,0], 1)
```

## Example 2
<img src="StudentLecture.png" alt="Drawing" style="width: 600px;"/>

```python
activity_names = ['lecture','web','hw','text']
A = np.array([
    [0.6, 0.4, 0.2, 0.3], # lecture
    [0.2, 0.5, 0.1, 0.2], # web
    [0.15,0.1, 0.7, 0.0], # hw
    [0.05,0.0, 0.0, 0.5]  # text
])
graph_matrix(A, mat_label=activity_names, show_weights=True, round_digits=3)
```

```python
x0 = np.array([0,0,0,1])
x1 = A@x0
x2 = A@x1
x3 = A@x2
x3
```

```python
its = 20
allx = np.zeros((4,its))

allx[:,0] = np.array([0, 0.2, 0.1, 0.7])

for i in range(1,its):
    allx[:,i] = A@allx[:,i-1]
    
allx.T
```

```python
plt.plot(allx.T)
plt.legend(activity_names)
plt.grid()
```

```python
lambdas, U = la.eig(A)
print(lambdas)
print(U[:,0])
```

```python
U[:,0] / la.norm(U[:,0], 1)
```

## Example 3

```python
name_mapping={0:'Google', 1:'Facebook', 2:'Amazon', 3:'wikipedia', 4:'Twitter', 5:'Youtube'}
```

```python
edges = [
    [0,1], [0,5],
    [1,2], [1,3],
    [2,3], [2,4], [2,5],
    [3,0],
    [4,5],
    [5,0]]

A = np.zeros((6,6))
for i,j in edges:
    A[j, i] = 1
    
graph_matrix(A, list(name_mapping.values()), show_weights=True)
```

```python
A
```

```python
np.sum(A,axis=0)
```

```python
M = A/np.sum(A,axis=0)
M
```

```python
graph_matrix(M, list(name_mapping.values()), show_weights=True)
```

```python
its = 20
allx = np.zeros((6,its))

allx[:,0] = np.array([1, 0, 0, 0, 0, 0])

for i in range(1,its):
    allx[:,i] = M@allx[:,i-1]
    
allx.T
```

```python
plt.plot(allx.T)
plt.legend(list(name_mapping.values()))
```

```python
edges = [
    [0,1], [0,5],
    [1,2], [1,3],
    [2,3], [2,4], [2,5],
    [3,0],
    [4,5]]

A = np.zeros((6,6))
for i,j in edges:
    A[j, i] = 1
    
graph_matrix(A, list(name_mapping.values()), show_weights=False)
```

```python
A
```

```python
A[:, -1] = 1
A
```

```python
M = A/np.sum(A,axis=0)
M
```

```python
graph_matrix(M, list(name_mapping.values()), show_weights=True)
```

```python
its = 20
allx = np.zeros((6,its))

allx[:,0] = np.array([1, 0, 0, 0, 0, 0])

for i in range(1,its):
    allx[:,i] = M@allx[:,i-1]
    
plt.plot(allx.T)
plt.legend(list(name_mapping.values()))
```
