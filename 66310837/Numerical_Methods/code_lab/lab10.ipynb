```python
import numpy as np
import numpy.linalg as la
import scipy.linalg as sla
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.exp(x) - 2

def df(x):
    return np.exp(x)
```

```python
x = np.linspace(-1, 2, 100)
plt.figure(figsize=(6,6))
plt.plot(x, f(x), lw=3)
```

```python
xhat = 1.0 # point where we want to find the approximation
h = 1.0 # initial perturbation
errors = []
hs = []

fval = f(xhat) # we only need to evaluate this once!

# in general, we don't have this value, but we will use it here to visualize the error
dfexact = df(xhat)

for i in range(20):
    
    # one function evalution per each perturbation size
    dfapprox = ( f(xhat+h) - fval ) / h
    
    # get the error 
    err = np.abs(dfexact - dfapprox)
    
    print(" %E \t %E " %(h, err) )
    
    hs.append(h)
    errors.append(err)
    
    h = h /2
```

```python
plt.figure(figsize=(16,10))
plt.loglog(hs, errors, lw=3)
plt.xlabel('h')
plt.ylabel('error')
plt.xlim([1e-6,1])
plt.ylim([1e-6,1])
plt.grid()
```

```python
plt.figure(figsize=(16,10))
plt.loglog(hs, errors, lw=3, label='total')
plt.loglog(hs, np.array(hs), '--', label='truncation')
plt.loglog(hs, 1e-16/np.array(hs), '--', label='rounding')
plt.legend(bbox_to_anchor=(1.1, 1.05))
plt.xlabel('h')
plt.ylabel('error')
plt.grid()
plt.ylim(1e-16,1e2)
```

## Riemanns Integral

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import collections as matcoll
```

```python
a = 0
b = np.pi
n = 101
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

lines = []
for i in range(len(x)):
    pair=[(x[i],0), (x[i], f[i])]
    lines.append(pair)
    
linecoll = matcoll.LineCollection(lines)
fig, ax = plt.subplots()
ax.add_collection(linecoll)

plt.plot(x,f)
```

```python
I_riemannL = h * sum(f[:n-1])
err_riemannL = 2 - I_riemannL

I_riemannR = h * sum(f[1:])
err_riemannR = 2 - I_riemannR

I_mid = h* sum(np.sin((x[:n-1] + x[1:])/2))
# x[:n-1] 0 1 2 ... 9
# x[1:]   1 2 3 ... 10 
err_mid = 2 - I_mid

print(I_riemannL)
print(err_riemannL)

print(I_riemannR)
print(err_riemannR)

print(I_mid)
print(err_mid)
```

## Trapezoid Rule

```python
a = 0
b = np.pi
n = 101
h = (b - a) / (n - 1)
x = np.linspace(a, b, n)
f = np.sin(x)

I_trap = (h)*np.sum((f[:n-1]+f[1:])/2)

err_trap = 2 - I_trap

print(I_trap)
print(err_trap)
```
